```python
import torchaudio
import torchaudio.transforms as transforms
import torch
import numpy as np
import matplotlib.pyplot as pl
```

```python
fname = "output/out_1.wav"

wav, sample_rate = torchaudio.load(fname, normalize=True)
transform = transforms.MelSpectrogram(sample_rate, n_fft=400, n_mels=128, hop_length=400)
mel_specgram = transform(wav).squeeze(0)
# log_spec = torch.clamp(mel_specgram, min=1e-10).log10()
# log_spec = torch.maximum(log_spec, log_spec.max() - 8.0)
# log_spec = (log_spec + 4.0) / 4.0
# pl.imshow(log_spec.detach().cpu().squeeze(0))
```
